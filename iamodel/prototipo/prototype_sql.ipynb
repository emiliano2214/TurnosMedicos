# Prototipo Generación de SQL

Este notebook prueba la capacidad del LLM para generar consultas SQL basándose en los ejemplos proporcionados.

```python
import os
from langchain_community.document_loaders import TextLoader
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from dotenv import load_dotenv

load_dotenv()
```

## 1. Cargar Contexto de SQL
Cargamos `ejemplos_sql.md` y `modelo_datos_oltp.md` para darle contexto al modelo.

```python
loader_sql = TextLoader("../docs/ejemplos_sql.md")
loader_model = TextLoader("../docs/modelo_datos_oltp.md")

docs_sql = loader_sql.load()
docs_model = loader_model.load()

context_text = "\n\n".join([d.page_content for d in docs_model + docs_sql])
print("Contexto cargado.")
```

## 2. Definir el Prompt
Instruimos al modelo para que actúe como un experto en SQL Server.

```python
system_prompt = """
Eres un asistente experto en SQL Server para el sistema TurnosMedicos.
Usa el siguiente contexto sobre el modelo de datos y ejemplos de consultas para responder.
Tu objetivo es generar una consulta SQL válida que responda a la pregunta del usuario.
Si no puedes generar una consulta segura, explica por qué.

Contexto:
{context}
"""

prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("user", "{question}")
])
```
